```python
import requests 
import json
import pandas as pd
import numpy as np
```

```python
Synonyms = pd.read_csv('Synonyms Dictionary.txt')
Synonyms = Synonyms.reset_index(drop=True)
SynonymsDictionary = Synonyms.set_index('Word')['Word2'].to_dict()

ProcessedPast = pd.read_csv('Processed AD+BE+CH+IL+L+PAN+PCP+PS.txt', index_col=False)
ProcessedPastSingleWords = ProcessedPast.query("Q_Words == 1")
UniqueWords = ProcessedPastSingleWords.Word.unique()
UniqueWords.sort()
```

```python
for Word in UniqueWords:
    if not(Word in SynonymsDictionary) and not(Word[0].isdigit()):
        print(Word)
        response = requests.get('https://dicionario.acad-ciencias.pt/pesquisa/?word=' + Word )
        #start = response.text.find('<div class=\'entry\'><div class=\'def\'><entry id="w1"')
        #start = response.text.find('xml:id="DLP-', start)+len('xml:id="DLP-')
        #end = response.text.find('_', start)
        start = response.text.find('Esta palavra não se encontra dicionarizada.')
        if start == -1:
            start = response.text.find('<div class=\'entry\'><div class=\'def\'><entry id="w1"')
            start = response.text.find('<orth', start)+len('<orth')
            start = response.text.find('>', start)+len('>')
            end = response.text.find('</orth>', start)

            #print(Word + "->")
            #print(start)
            #print(end)
            #print(response.text)
            WordSynonym = response.text[start:end].strip()
            #print("->" + WordSynonym)
        else:
            WordSynonym = ""
        #if Word != WordSynonym:
        Synonyms = Synonyms.append({"Word": Word, "Word2": WordSynonym}, ignore_index=True)
```
